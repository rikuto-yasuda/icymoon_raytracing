```python
cdf_file = 'C:\\Users\\tsuch\\Dropbox\\JUICE_GS\\Data\\2022_NECP_runthrough\\Output\\RPWI_S8_270_2022_06_28_18_16_12\\JUICE_L1a_RPWI-HF-SID3_20000101T005727-20000101T010715_V01___TMIDX_00000.bin.cdf'
cdf = pycdf.CDF(cdf_file)
scet = cdf['SCET'][...]
freq = cdf['frequency'][...]
power_of_x = cdf['power_of_x'][...]
power_of_y = cdf['power_of_y'][...]
power_of_z = cdf['power_of_z'][...]
```

```python
n = len(power_of_x)
n_freq = 255
n_time = int(n/255)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)
ax1.plot(power_of_x, '-r')
ax1.plot(power_of_y, '-g')
ax1.plot(power_of_z, '-b')
ax1.set_yscale('log')
ax2.plot(freq)
ax3.plot(scet)
ax3.set_xlabel('data number')
ax1.set_ylabel('Power')
ax2.set_ylabel('Freq[kHz]')
ax3.set_ylabel('SCET')
fig.show
```

```python
freq_2d    = freq.reshape([n_time, n_freq])
scet_2d    = scet.reshape([n_time, n_freq])
power_x_2d = power_of_x.reshape([n_time, n_freq]).transpose()
power_y_2d = power_of_y.reshape([n_time, n_freq]).transpose()
power_z_2d = power_of_z.reshape([n_time, n_freq]).transpose()
```

```python
freq_1d = freq_2d[0,:]
scet_1d = scet_2d[:,0]
```

```python
fig2d = plt.figure(figsize=[10,10])
ax1 = fig2d.add_subplot(3, 1, 1)
ax2 = fig2d.add_subplot(3, 1, 2)
ax3 = fig2d.add_subplot(3, 1, 3)
p1=ax1.pcolormesh(scet_1d, freq_1d, np.log10(power_x_2d))
p2=ax2.pcolormesh(scet_1d, freq_1d, np.log10(power_y_2d))
p3=ax3.pcolormesh(scet_1d, freq_1d, np.log10(power_z_2d))
ax1.set_title('power of X')
ax2.set_title('power of Y')
ax3.set_title('power of Z')
ax3.set_xlabel('SCET')
ax1.set_ylabel('Freq[kHz]')
ax2.set_ylabel('Freq[kHz]')
ax3.set_ylabel('Freq[kHz]')
pp1 = fig.colorbar(p1, ax=ax1, orientation="vertical")
pp1.set_label('Log power')
pp2 = fig.colorbar(p2, ax=ax2, orientation="vertical")
pp2.set_label('Log power')
pp3 = fig.colorbar(p3, ax=ax3, orientation="vertical")
pp3.set_label('Log power')

plt.show()
```
